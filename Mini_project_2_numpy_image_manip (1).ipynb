# Mini_project_2 : Manipulating images using Numpy

```python
import matplotlib.pyplot as plt
import skimage.io
import numpy as np
from PIL import Image
```

# Loading image to numpy array and displaying it

```python
def load():
    image = skimage.io.imread(fname="panda.jpg")
    fig, ax = plt.subplots()
    plt.imshow(image)
    plt.show()
    print("image loaded successfully!!")
```

# horizontally flipping original image

```python
def horiz_flip():
    horizontal_img=image[:, ::-1] 
    plt.imshow(horizontal_img)
    plt.show()
```

# vertically flipping original image

```python
def vert_flip():
    vertical_img=image[::-1]
    plt.imshow(vertical_img)
    plt.show()
```

# Inverting image colour

```python
def invert_color():
    invert_img=image[:, :, ::-1]
    plt.imshow(invert_img)
    plt.show()
```

# converting to grayscale

```python
def imgtogray(image):
    gray=np.mean(image, axis=-1)
    return gray
```

# converting to Black & White

```python
def Convert_to_BW(image):
    bw=imgtogray(image)
    np_image_blackandwhite = np.copy(bw)
    np_image_blackandwhite[np_image_blackandwhite < 128] = 0
    np_image_blackandwhite[np_image_blackandwhite >= 128] = 255
    return np_image_blackandwhite    
```

# saving the image 

```python
def save():
    im = Image.fromarray(image)
    im.save("pandas_saved.jpeg")
    print("image saved successfully!!")
```

# main function

```python
#MAIN FUNCTION
if __name__=="__main__":
    image = skimage.io.imread(fname="panda.jpg")
    choice = None
    while choice != "0":
        print \
            ("""
                ---WELCOME---
                0. Quit
                1. Load Image
                2. Edit Image
                3. Save Image
                
                """)

        choice = input("Your choice:")
        print()
        if choice == "1":
            load()
        elif choice == "2":  # when user chooses to edit image
            print('Welcome  \n----Choose preferred options:----')

            print('0. Horizontal Flip\n1. Vertical Flip\n2. Invert Colors\n3. Convert to Grayscale\n4. Convert to B&W\n5.Exit')

            user_choice = int(input('Enter choice:'))
            if user_choice == 0:  # when user chooses Horizontal Flip
                horiz_flip()

            elif user_choice == 1:  # when user chooses Vertical Flip
                vert_flip()
                    
            elif user_choice == 2:  # when user chooses to Invert Colors
                invert_color()
                
            elif user_choice == 3:  # when user chooses to Convert to Grayscale
                gray_image=imgtogray(image)
                plt.imshow(gray_image,cmap="gray")
                plt.show()

            elif user_choice == 4:  # when user chooses to Convert to B&W
                bw_img=Convert_to_BW(image)
                plt.imshow(bw_img, cmap='gray')
                plt.show()
            else:  # when user choose exit or enter wrong key
                print("Session expired")

        elif choice == "3": 
            save()
        
        elif choice == "0":  # when user choose exit
            print("Thank you!")
```
